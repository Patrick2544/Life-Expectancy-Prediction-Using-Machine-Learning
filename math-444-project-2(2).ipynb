```r
library(dplyr)

data_no_country <- data %>% select(-Country)
data_no_country$Status <- as.factor(data_no_country$Status)
```

```r
# Impute using missForest
library(missForest)
imputed_data <- missForest(data_no_country, maxiter = 10, ntree = 100)

# Summary statistics after imputation
summary(imputed_data$ximp)
```

```r
library(missForest)

# Tune the parameters
optimized_imputed_data <- missForest(
  data_no_country, 
  maxiter = 15,         # Increase maximum iterations for better refinement
  ntree = 200,          # Increase the number of trees for more accurate imputations
  mtry = floor(sqrt(ncol(data_no_country)))  # Use default heuristic for mtry
)

# View the updated OOB imputation error
optimized_imputed_data$OOBerror
```

```r
# Convert the imputed data back to a data frame
imputed_data <- as.data.frame(optimized_imputed_data$ximp)

# Add the Country column from the original dataset
imputed_data_with_country <- cbind(imputed_data, Country = data$Country)
```

```r
# Split the data based on the Year
train_split <- imputed_data_with_country %>% filter(Year == 2014)
test_split <- imputed_data_with_country %>% filter(Year == 2015)

# Verify the splits
cat("Number of rows in training set (2014):", nrow(train_split), "\n")
cat("Number of rows in test set (2015):", nrow(test_split), "\n")

# Save the new train and test datasets
write.csv(train_split, "imputed_train_data_2014.csv", row.names = FALSE)
write.csv(test_split, "imputed_test_data_2015.csv", row.names = FALSE)
```

```r
imputed_test <- read.csv("/kaggle/working/imputed_test_data_2015.csv")

imputed_train <- read.csv("/kaggle/working/imputed_train_data_2014.csv")
```

```r
dim(imputed_test)
dim(imputed_train)
```

```r
formula <- Life.expectancy ~ Status + Adult.Mortality + infant.deaths + 
  Alcohol + percentage.expenditure + Hepatitis.B + Measles + 
  BMI + under.five.deaths + Polio + Total.expenditure + Diphtheria + 
  HIV.AIDS + GDP + Population + thinness..1.19.years + thinness.5.9.years + 
  Income.composition.of.resources + Schooling
```

```r
full_model <- lm(formula, data = imputed_train)

summary(full_model)
```

```r
plot(full_model)
```

```r
library(MASS)
```

```r
# Perform stepwise selection
stepwise_model <- stepAIC(full_model, scope = formula, direction = "both")
```

```r
null_model <- lm(Life.expectancy ~ 1, data = imputed_train)

forward_model <- stepAIC(null_model, scope = formula, direction = "forward")
```

```r
backward_model <- stepAIC(full_model, scope = formula, direction = "forward")
```

```r
summary(forward_model)
```

```r
plot(forward_model)
```

```r
boxcox_result <- boxcox(forward_model, lambda = seq(-2, 2, by = 0.1))

# Find the optimal lambda
optimal_lambda <- boxcox_result$x[which.max(boxcox_result$y)]
cat("Optimal lambda:", optimal_lambda, "\n")
```

```r
# Apply transformation with the optimal lambda
if (optimal_lambda != 0) {
  imputed_train$Life.expectancy_transformed <- 
    (imputed_train$Life.expectancy^optimal_lambda - 1) / optimal_lambda
} else {
  imputed_train$Life.expectancy_transformed <- log(imputed_train$Life.expectancy)
}

# Refit the model with the transformed response variable
lm_fit_transformed <- lm(Life.expectancy_transformed ~ Income.composition.of.resources + 
                            HIV.AIDS + Adult.Mortality + Total.expenditure, 
                          data = imputed_train)

# Summarize the new model
summary(lm_fit_transformed)
```

```r
plot(lm_fit_transformed)
```

```r
# Ensure the same transformation is applied to the test response variable, if applicable
if (optimal_lambda != 0) {
  imputed_test$Life.expectancy_transformed <- 
    (imputed_test$Life.expectancy^optimal_lambda - 1) / optimal_lambda
} else {
  imputed_test$Life.expectancy_transformed <- log(imputed_test$Life.expectancy)
}

# Make predictions using the transformed model
predicted_values <- predict(lm_fit_transformed, newdata = imputed_test)

# Calculate RMSE
actual_values <- imputed_test$Life.expectancy_transformed
rmse <- sqrt(mean((predicted_values - actual_values)^2))
cat("RMSE:", rmse, "\n")
```
